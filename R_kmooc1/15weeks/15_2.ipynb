```python
i = 0
A1 = []
A2 = []
A3 = []
A3
type(A3)
```

```python
while i < len(fpop) : #fpop 행 개수 
    temp = pname[i].split()
    A1.insert(i, temp[0])
    A2.insert(i, temp[1])
    A3.insert(i, temp[2])
    i = i + 1
```

```python
fpop['A1'] = A1
fpop['A2'] = A2
fpop['A3'] = A3
```

```python
i = 0
TOTAL = []
WOMEN = []
MEN = []
```

```python
while i < len(그룹핑_fpop) :
    each_MEN = 그룹핑_fpop.iloc[i]['M10'] + 그룹핑_fpop.iloc[i]['M20'] + 그룹핑_fpop.iloc[i]['M30'] + 그룹핑_fpop.iloc[i]['M40'] + 그룹핑_fpop.iloc[i]['M50'] + 그룹핑_fpop.iloc[i]['M60']
    MEN.insert(i,each_MEN)
    
    each_WOMEN = 그룹핑_fpop.iloc[i]['W10'] + 그룹핑_fpop.iloc[i]['W20'] + 그룹핑_fpop.iloc[i]['W30'] + 그룹핑_fpop.iloc[i]['W40'] + 그룹핑_fpop.iloc[i]['W50'] + 그룹핑_fpop.iloc[i]['W60']
    WOMEN.insert(i,each_WOMEN)
    
    each_TOTAL = each_MEN + each_WOMEN
    TOTAL.insert(i,each_TOTAL)
    
    i=i+1
```

```python
그룹핑_fpop['MEN'] = MEN
그룹핑_fpop["WOMEN"] = WOMEN
그룹핑_fpop["TOTAL"] = TOTAL
그룹핑_fpop
```

```python
그룹핑_fpop.columns
```

```python
그룹핑_fpop_new = 그룹핑_fpop[['M10','M20','M30','M40','M50','M60','W10','W20','W30','W40','W50','W60','MEN','WOMEN','TOTAL']]
```
